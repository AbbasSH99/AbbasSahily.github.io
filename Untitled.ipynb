```python
import pandas as pd
from sqlalchemy import create_engine
import os
from glob import glob

# Database connection parameters
server = 'DESKTOP-1512MV9\\SQLEXPRESS'
database = 'COD_team_correlation'
username = 'DESKTOP-1512MV9\\ACER'
driver = 'ODBC Driver 17 for SQL Server'

# Create a connection string (no password needed)
connection_string = f'mssql+pyodbc://{username}@{server}/{database}?driver={driver}'
engine = create_engine(connection_string)

# Path to the directory containing the Excel files
file_path = r'C:\Users\ACER\Desktop\Upwork_Projects\cod_data_project\table_by_player'

# Load all Excel files
all_files = glob(os.path.join(file_path, "*.xlsx"))

# Function to load and upload each Excel file to SQL Server
def load_and_upload(file, table_name):
    df = pd.read_excel(file)
    # If your data requires specific handling, you can add more data cleaning steps here
    df.to_sql(table_name, con=engine, if_exists='append', index=False)
    print(f'Uploaded {file} to table {table_name}')

# Loop through each file and upload
for file in all_files:
    # Extract table name from file name (modify as necessary)
    table_name = os.path.splitext(os.path.basename(file))[0]
    load_and_upload(file, table_name)

print("All files have been uploaded successfully.")
```
